```python
#to load the datasets from sns
df=sns.load_dataset('titanic')
df.head()
```

```python
df=sns.load_dataset('exercise')
df
```

```python
#COUNTPLOT
#is useful when dealing with the cataborical values and used to plot the frequency of diff. catagories
```

```python
sns.countplot(x="diet",data=df)
```

```python
sns.countplot(x="sex",hue="survived",data=df)
```

```python
sns.countplot(x="sex",hue="survived",data=df,palette='Set2')
```

```python
#HEATMAP
#will give co relations of the columns in terms of color
```

```python
df1=sns.load_dataset("iris")
df1
```

```python
corr=df1.corr()
sns.heatmap(corr)
```

```python
#KDE PLOT
# this gives distribution
```

```python
sns.kdeplot(x="age",data=df,color='black')
```

```python
#DISPLOT
#can check distribution of continous datas
```

```python
sns.displot(x='age',kde=True,bins = 5, data=df)
```

```python
sns.displot(x='age',kde=False,bins = 5, data=df)
```

```python
sns.displot(x='age',kde=True,bins = 5,hue=df['survived'], data=df) #we use hue to distribute more 
```

```python
sns.displot(x='sex',kde=True,bins = 5,hue=df['survived'], data=df)
```

```python
sns.displot(x='age',kde=True,bins = 5,hue=df['survived'], data=df,palette='Set1')
```

```python
#SCATTERPLOT
```

```python
sns.scatterplot(x ='sepal_length', y ='petal_length',data=df1)
```

```python
sns.scatterplot(x ='sepal_length', y ='petal_length',data=df1,hue='species') #hue to distribute species too
```

```python
#pair - it gives relation between all the plots
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df1)
```

```python
#joint plot
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=df1)
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=df1, kind='reg') #reg-regression
```

```python
sns.jointplot(x='sepal_length',y='petal_length',data=df1,kind='hex')
```
